```python
#If you want to run this, you'll have to chage the file path. I used Parks.csv and Teams.csv
df = pd.read_csv('/Users/matthewcoker/Documents/College/2018Spring/INFO/FinalProject/spring-2018-final-projects-matts-team/baseballdatabank-master/core/Teams.csv')
parksDF = pd.read_csv('/Users/matthewcoker/Documents/College/2018Spring/INFO/FinalProject/spring-2018-final-projects-matts-team/baseballdatabank-master/core/Parks.csv')
```

```python
df.fillna(0, inplace = True)
```

```python
dfTeams = df[['yearID','teamID','park','HR','G']]
allParks = dfTeams['park'].unique()
hrAvgPerPark = []
parks = []
for park in allParks:
    hrAvg = 0
    givenPark = dfTeams.loc[dfTeams['park']==park]
    games = givenPark['G'].sum()
    smackers = givenPark['HR'].sum()
    if games >= 5:
        hrAvg = smackers/games
    else:
        hrAvg = 0
    hrAvgPerPark.append(hrAvg)
```

```python
data = pd.DataFrame({'Parks':allParks, 'HR Average':hrAvgPerPark})
pd.set_option('display.max_rows', None)
display(data)
```
